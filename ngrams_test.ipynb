```python
from scipy import sparse
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors
```

```python
df = pd.read_csv("base.csv", sep=",", names=["names", "link", "text"])[
    ["names", "link", "text"]
]
df_1 = pd.read_csv("to_search_for.csv", sep=",", names=["text"])["text"]
```

```python
samples = df["text"]

vectorizer = TfidfVectorizer(ngram_range=(2,2), min_df=1, max_df=1, token_pattern=r"\b\w+\b|\S*\d+")
X_train_tfidf = vectorizer.fit_transform(samples)

neigh = NearestNeighbors(n_neighbors=20, n_jobs=-1)
neigh.fit(X_train_tfidf)
```

```python
ll=[]
test=df_1
for k in test:
    X_tfidf2 = vectorizer.transform([k])
    ll.append(X_tfidf2)

ll = sparse.vstack((ll))
nn = 10
res = neigh.kneighbors(ll, n_neighbors=nn ,return_distance=True)
```
